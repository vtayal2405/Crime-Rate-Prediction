```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import metrics
```

```python
 dataset = pd.read_excel("/content/crp.xlsx", sheet_name="Sheet1")
dataset.head()
```

```python
dataset.info()
```

```python
fig, ax = plt.subplots(11, 1, figsize=(10, 50))

for i in range(0, 11):
    ax[i].barh(dataset['City'], dataset[dataset.columns[i+2]], 0.6, color='Salmon')
    ax[i].set_title('City vs ' + dataset.columns[i+2])
plt.show()
```

```python
new_df = pd.DataFrame(columns=['Year', 'City', 'Population (in Lakhs) (2011)+', 'Number Of Cases', 'Type'])
for i in range(3, 13):
    temp = dataset[['Year', 'City', 'Population (in Lakhs) (2011)+']].copy()
    temp['Number Of Cases'] = dataset[[dataset.columns[i]]]
    temp['Type'] = dataset.columns[i]

    new_df = pd.concat([new_df, temp])
new_df
```

```python
new_df['Crime Rate'] = new_df['Number Of Cases'] / new_df['Population (in Lakhs) (2011)+']
new_df
```

```python
new_df = new_df.drop(['Number Of Cases'], axis=1)
new_df
```

```python
new_df.to_excel("/content/new_dataset.xlsx", index=False, sheet_name ='Sheet1')
```

```python
new_dataset = pd.read_excel("/content/new_dataset.xlsx", sheet_name="Sheet1")
new_dataset
```

```python
new_dataset.info()
```

```python
new_dataset.describe()
```

```python
le = LabelEncoder()
new_dataset['City'] = le.fit_transform(new_dataset['City'])
mapping = dict(zip(le.classes_, range(len(le.classes_))))
# Saving the mapping file for further use
file = open('/content/City_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python
new_dataset['Type'] = le.fit_transform(new_dataset['Type'])
mapping = dict(zip(le.classes_, range(len(le.classes_))))
# Saving the mapping file for further use
file = open('/content/Type_Mapping.txt', 'wt')
for key,val in mapping.items():
    print(str(key) + " - " + str(val) + '\n')
    file.write(str(key) + " - " + str(val) + '\n')
```

```python
x = new_dataset[new_dataset.columns[0:4]].values
x
```

```python
y = new_dataset['Crime Rate'].values
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=50)
x_train
```

```python
y_train
```

## **Support** **Vector** **Machine**

```python
model1 = svm.SVR()
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
result1=metrics.r2_score(y_test, y_pred)
print('R2 score:', result1)
```

## **Nearest** **Neighbour**

```python
model2 = KNeighborsRegressor(n_neighbors=2)
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
result2=metrics.r2_score(y_test, y_pred)
print('R2 score:', result2)
```

## **Decision Tree Regressor**

```python
model3 = tree.DecisionTreeRegressor()
model3.fit(x_train, y_train)
y_pred = model3.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
result3=metrics.r2_score(y_test, y_pred)
print('R2 score:', result3)
```

## **Random Forest Regressor**

```python
model4 = RandomForestRegressor(random_state=0)
model4.fit(x_train, y_train)
y_pred = model4.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
result4=metrics.r2_score(y_test, y_pred)
print('R2 score:', result4)
```

## **Neural Networks MLPRegressor**

```python
model5 = MLPRegressor(random_state=0)
model5.fit(x_train, y_train)
y_pred = model5.predict(x_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
result5=metrics.r2_score(y_test, y_pred)
print('R2 score:', result5)
```

```python
data={"Support Vector Machine":result1,"Nearest Neighbour":result2,"Decision Tree Regressor":result3,
      "Random Forest Regressor":result4,"Neural Networks MLPRegressor":result5}
Algorithm = list(data.keys())
R2_value = list(data.values())

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(Algorithm, R2_value, color ='maroon',
        width = 0.7)
plt.xticks(rotation=45)
plt.xlabel("Algorithms used")
plt.ylabel("R2 Value")
plt.title("Accuracy")
plt.show()
```
